#Load Data

```python
#Bước 3: load dữ liệu dân số theo quận:
import geopandas as gpd

filename = "/content/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp"

df_pop = gpd.read_file(filename)

df_pop
```

```python
#Bước 4: Load dữ liệu click của người dùng
df_click = gpd.read_file("/content/drive/MyDrive/CS116/response3.json")

df_click
```

```python
backup_crs = df_click.crs
df_click = df_click.to_crs(df_pop.crs)
```

#Lấy top 10 quận/huyện có tốc độ dân số lớn nhất

```python
#Bước 5: Lấy ra top 10 quận/huyện có tốc độ tăng dân số lớn nhất
grow_rate = df_pop[['Dist_Name', 'Pop_2017', 'Pop_2019']]
grow_rate['speed'] = grow_rate['Pop_2019']/grow_rate['Pop_2017']
top_rate = grow_rate.nlargest(10, 'speed')['Dist_Name']
top_rate
```

```python
#lấy list tên các quận huyện đó
import numpy as np
dist_name = np.array(top_rate)
dist_name
```

#Lọc lấy các click thuộc 10 quận huyện vừa mới lọc

```python
#Bước 6: Lọc các điểm click theo top 10 quận huyện này
df_point = gpd.sjoin(df_click, df_pop, how='left', op='within')
top10_speed = df_point[df_point['Dist_Name'].isin(dist_name)][['geometry', 'Dist_Name']]
top10_speed = top10_speed.to_crs(backup_crs)
```

```python
top10_speed
```

#Dùng K_Means để khoanh vùng các nơi click.

```python
#Bước 7: Chạy thuật toán gom nhóm K-Means với K = 20
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
def get_kmean(df):
  point_x = df['geometry'].apply(lambda dt: dt.x)
  point_y = df['geometry'].apply(lambda dt: dt.y)
  X = np.column_stack((point_x, point_y))
  kmeans = KMeans(n_clusters=4)
  kmeans.fit(X)
  df['cluster'] = kmeans.labels_
  return df
```

#Lấy các cụm điểm cho từng cụm

```python
#Bước 8: Lưu cụm điểm nhiều nhất của mỗi quận
data_lst = []
for name in list(dist_name):
  df = get_kmean(top10_speed[top10_speed['Dist_Name'] == name])
  df_max = df[df['cluster'] == int(df.cluster.mode())]
  data_lst.append(df_max)
```

```python
frame = pd.concat(data_lst)
frame
```

#Vẽ biểu đồ HeatMap

```python
#Bước 9: Vẽ biểu đồ heatmap với thư viện folium các cụm điểm lớn nhất theo từng quận/huyện này
import folium
from folium.plugins import HeatMap

map = folium.Map(location=[10.9, 106.54], zoom_start=10, tiles='CartoDB positron')
map
```

```python
lon = np.array(frame['geometry'].apply(lambda dt: dt.x))
lat = np.array(frame['geometry'].apply(lambda dt: dt.y))
data = np.vstack((lat, lon)).T
HeatMap(data).add_to(folium.FeatureGroup(name='Heat Map').add_to(map))
folium.LayerControl().add_to(map)
```
